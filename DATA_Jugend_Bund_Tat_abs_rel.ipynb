### Create DataFrame that can be used for EDA juvenile crime:
* delete columns with age groups we don´t need
* summarize age groups
* filter for keys "schluessel" we decided to look at

```python
df_1 = bund_tat_all.copy()
columns_df_1 = df_1.columns.tolist()
columns_df_1
```

```python
df_2 = df_1[['schluessel', 'straftat', 'sexus', 'tatverdaechtige_insgesamt','kinder_10_bis_unter_12', 'kinder_12_bis_unter_14', 'kinder_unter_14', 'jugendliche_14_bis_unter_16', 'jugendliche_16_bis_unter_18', 'jugendliche_14<18', 'heranwachsende_18_bis_unter_21', 'erwachsene_ab_21', 'year']]
df_2.head(3)
```

```python
# add 2 age groups:
df_2["kinder_10_bis_unter_14"] = df_2["kinder_10_bis_unter_12"] + df_2["kinder_12_bis_unter_14"]
df_2["jugendl_u_heranwachsende_14_bis_unter_21"] = df_2["jugendliche_14<18"] + df_2["heranwachsende_18_bis_unter_21"]

# bring columns in correct order:
df_3 = df_2[['schluessel', 'straftat', 'sexus', 'tatverdaechtige_insgesamt','kinder_10_bis_unter_12', 'kinder_12_bis_unter_14', 'kinder_10_bis_unter_14', 'kinder_unter_14', 'jugendliche_14_bis_unter_16', 'jugendliche_16_bis_unter_18', 'jugendliche_14<18', 'heranwachsende_18_bis_unter_21', 'jugendl_u_heranwachsende_14_bis_unter_21', 'erwachsene_ab_21', 'year']]
df_3.head()
```

```python
ausgew_keys = ["------", "010000", "020010", "220000", "100000", "232100", "232200", "435*00", "*26*00", "210000", "730000", "891100", "674000"]
```

```python
df_4 = df_3[df_3["schluessel"].isin(ausgew_keys)]
df_4.reset_index(inplace=True)
df_4.drop(columns="index", inplace=True)
df_4.info()
```

```python
df_4["year"] = df_4["year"].astype("int")
```

### aggregating "Mord" und "Totschlag" and adding equivalent rows:

```python
agg_keys = ["010000", "020010"]

keys_as_string = ""
for i in agg_keys:
    j = i + ", "
    keys_as_string += j
df_5 = df_4[df_4["schluessel"].isin(agg_keys)]
keys_as_string = keys_as_string[0:-2]

gender = ["M", "W", "X"]
year = [2018, 2019, 2020, 2021, 2022]
row_labels = []

df_6 = df_5.copy()
        
for y in year:
    for g in gender:
        row_label = g + str(y)
        df_6.loc[row_label] = df_6[(df_6["sexus"] == g) & (df_6["year"] == y)].sum()
        df_6.loc[row_label, "schluessel"] = keys_as_string
        df_6.loc[row_label, "sexus"] = g
        df_6.loc[row_label, "year"] = y
        row_labels.append(row_label)

for l in row_labels:
    df_4.loc[l] = df_6.loc[l]

#drop Mord und Totschlag seperate
df_4 = df_4[df_4['schluessel'].isin(["------", "010000, 020010", "220000", "100000", "232100", "232200", "435*00", "*26*00", "210000", "730000", "891100", "674000"])]

df_4.reset_index(inplace=True)
df_4.drop(columns="index", inplace=True)

df_4.head()
```

```python
bund_jugend_tat_absolut = df_4.copy()
bund_jugend_tat_absolut.tail()
```
