```python
# add 2 age groups:
df_3["kinder_10_bis_unter_14"] = df_3["kinder_10_bis_unter_12"] + df_3["kinder_12_bis_unter_14"]
df_3["jugendl_u_heranwachsende_14_bis_unter_21"] = df_3["jugendliche_14<18"] + df_3["heranwachsende_18_bis_unter_21"]

# bring columns in correct order:
df_4 = df_3[['schluessel', 'straftat', 'bundesland', 'sexus', 'tatverdaechtige_insgesamt', 'kinder_10_bis_unter_12', 'kinder_12_bis_unter_14', 'kinder_10_bis_unter_14', 'kinder_unter_14', 'jugendliche_14_bis_unter_16', 'jugendliche_16_bis_unter_18', 'jugendliche_14<18', 'heranwachsende_18_bis_unter_21', 'jugendl_u_heranwachsende_14_bis_unter_21', 'erwachsene_ab_21', 'year']]
df_4.columns
```

```python
ausgew_keys = ["------", "010000", "020010", "220000", "100000", "232100", "232200", "435*00", "*26*00", "210000", "730000", "891100", "674000"]
```

```python
df_5 = df_4[df_4["schluessel"].isin(ausgew_keys)]
df_5.reset_index(inplace=True)
df_5.drop(columns="index", inplace=True)
```

```python
df_5.head(10)
```

```python
bundeslaender = df_5["bundesland"].unique().tolist()
bundeslaender
```

### aggregating "Mord" und "Totschlag" and adding equivalent rows:

```python
agg_keys = ["010000", "020010"]

keys_as_string = ""
for i in agg_keys:
    j = i + ", "
    keys_as_string += j
df_6 = df_5[df_5["schluessel"].isin(agg_keys)]
keys_as_string = keys_as_string[0:-2]

gender = ["M", "W", "X"]
year = [2018, 2019, 2020, 2021, 2022]
row_labels = []

df_7 = df_6.copy()
        

for y in year:
    for g in gender:
        for b in bundeslaender:
            row_label = g + str(y) + b
            df_7.loc[row_label] = df_7[(df_7["bundesland"] == b) & (df_7["sexus"] == g) & (df_7["year"] == y)].sum()
            df_7.loc[row_label, "schluessel"] = keys_as_string
            df_7.loc[row_label, "bundesland"] = b
            df_7.loc[row_label, "sexus"] = g
            df_7.loc[row_label, "year"] = y
            row_labels.append(row_label)

for l in row_labels:
    df_5.loc[l] = df_7.loc[l]

df_5 = df_5[df_5["schluessel"].isin(["------", "010000, 020010", "220000", "100000", "232100", "232200", "435*00", "*26*00", "210000", "730000", "891100", "674000"])]


df_5.reset_index(inplace=True)
df_5.drop(columns="index", inplace=True)

df_5.tail(50)
```

```python
laender_jugend_tat_absolut = df_5.copy()
laender_jugend_tat_absolut.head()
```
